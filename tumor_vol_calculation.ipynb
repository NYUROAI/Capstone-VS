```python
#volume calcuation
import os
patient_tumor_vol = dict()
for file in Path(processed_dir).iterdir():
    for file2 in sorted(os.listdir(Path(file))):
        if file2.split('/')[-1] in all_VS_file_names:

            VSpt=tio.Subject(VSmask=tio.LabelMap(str(file) + '/' + str(file2)))
            unit_size = np.prod(VSpt.VSmask.spacing)
            vol = VSpt.VSmask.data[0].sum()/255*unit_size # unit: mm^3
            patient_tumor_vol[str(file) + '/' + str(file2)] = vol
            break
```

```python
pd.DataFrame.from_dict(patient_tumor_vol, orient= 'index', columns=['vol']).to_csv('patient_tumor_vol.csv')
```

```python
def image_range(orientation,sample):
    #orientation can only be axslice,corslice,sagslice
    #data format VSpt['VSmask']
    checker = True
    if orientation == 'axslice':
        for i in range(sample.shape[3]):
            if sample.data[0,:,:,i].sum() != 0:
                if checker:
                    start = i
                checker = False
            if (sample.data[0,:,:,i].sum() == 0) and (not checker) and (i > start) :
                end = i-1
                break
        return start,end  
    if orientation == 'corslice':
        for i in range(sample.shape[2]):
            if sample.data[0,:,i,:].sum() != 0:
                if checker:
                    start = i
                checker = False
            if (sample.data[0,:,i,:].sum() == 0) and (not checker) and (i > start) :
                end = i-1
                break
        return start,end  
    if orientation == 'sagslice':
        for i in range(sample.shape[1]):
            if sample.data[0,i,:,:].sum() != 0:
                if checker:
                    start = i
                checker = False
            if (sample.data[0,i,:,:].sum() == 0) and (not checker) and (i > start) :
                end = i-1
                break
        return start,end  
```

```python
#volume calcuation
import os
patient_tumor_vol = dict()
for file in Path(processed_dir).iterdir():
    ID_num = str(file).split('/')[-1]
    for file2 in sorted(os.listdir(Path(file))):
        try:
            if ground_truth_tumor_vol.loc[ID_num].target_name in file2:
                VSpt=tio.Subject(VSmask=tio.LabelMap(str(file) + '/' + str(file2)))
                unit_size = np.prod(VSpt.VSmask.spacing)
                vol = VSpt.VSmask.data[0].sum()/255*unit_size # unit: mm^3
                patient_tumor_vol[str(file) + '/' + str(file2)] = vol
                break
        except KeyError:
            continue
```
