```python
import numpy as np
import pandas as pd
import json
from sklearn.model_selection import train_test_split
```

```python
subjects = pd.read_excel('VS_censor_send.xlsx', sheet_name = 'Include')
```

```python
subjects
```

```python
y = subjects['target_name']
X = subjects.drop(columns = ['target_name'])
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.6)
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5)
```

```python
X_train
```

```python
subject = {}
```

```python
subject['training'] = np.array(X_train['ID'].str.lstrip('VS')).tolist()
```

```python
subject['validation'] = np.array(X_valid['ID'].str.lstrip('VS')).tolist()
```

```python
subject['test'] = np.array(X_test['ID'].str.lstrip('VS')).tolist()
```

```python
subject
```

```python
with open('model/VS_subject.json', 'w') as outfile:
    json.dump(subject, outfile)
```
